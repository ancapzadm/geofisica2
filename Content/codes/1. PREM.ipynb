# Preliminary Reference Earth Model (PREM)

O PREM é um modelo unidimensional que representa as propriedades médias da Terra em função do raio planetário. Este modelo inclui a profundidade, densidade, velocidades sísmicas, atenuação e o grau de anisotropia nos limites das camadas que compõem o planeta. 

Os dados serão baixados de uma biblioteca desenvolvida em Python chamada RockHound. Esta biblioteca possui uma série de conjunto de dados úteis para as mais diversas aplicações, todas em um só lugar. Ela faz parte de um projeto chamado [Fatiando a Terra](https://www.fatiando.org).

**Referências**:

* Dziewonski, A.M., and D.L. Anderson. 1981. *Preliminary reference Earth model*. Phys. Earth Plan. Int. 25:297-356.

* Uieda, L., and Soler, S.R. (2019). *Rockhound: Download geophysical models/datasets and load them in Python (Version 0.1.0)*. Zenodo. https://doi.org/10.5281/zenodo.3086002

## Importando bibliotecas

```python
import rockhound as rh
import matplotlib.pyplot as plt
```

### Baixando os dados

```python
prem = rh.fetch_prem()
```

```python
print (prem)
```

### Visualizando os dados de densidade e velocidades sísmicas

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 10), sharey=True)
fig.suptitle("PREM: Preliminary Reference Earth Model")
ax = axes[0]
prem.plot("density", "depth", legend=False, ax=ax)
ax.invert_yaxis()
ax.set_xlabel("Densidade [g/cm³]")
ax.set_ylabel("Profundidade [km]")
ax.grid()
ax = axes[1]
for velocity in ["Vpv", "Vsv"]:
    prem.plot(velocity, "depth", legend=False, ax=ax, label=velocity)
ax.grid()
ax.legend()
ax.set_xlabel("Velocity [km/s]")
plt.show()
```
